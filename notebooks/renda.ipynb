```python
import urllib3
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
```

```python
request = requests.get(f'https://servicodados.ibge.gov.br/api/v3/agregados/7437/periodos/2013|2014|2015|2016|2017|2018|2019|2020|2021/variaveis/10750?localidades=N3[all]&classificacao=11308[49204,49205]')
resultado = request.json()

lista_resultados = []

for serie in resultado[0]['resultados'][0]['series']:
        codigo_local = serie['localidade']['id']
        nome_local = serie['localidade']['nome']
        ano = list(serie['serie'].keys())
        for y in ano:
          valor = serie['serie'][y]
          lista_resultados.append(
              {
                'codigo_local':codigo_local,           
                'nome_local':nome_local,
                'ano':y,
                'valor': valor
              })
df = pd.DataFrame(lista_resultados)
df
```
